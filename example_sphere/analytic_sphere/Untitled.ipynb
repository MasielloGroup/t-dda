```python
import numpy as np
import matplotlib.pyplot as plt
from single_sphere_analytics import DipoleParameters
```

```python
wspan=np.arange(2, 3,.008)
rad_sph = 10.E-7 # cm
hbar_eVs = 6.58212E-16 # Planck's constant [eV*s]     
sphere = DipoleParameters(radius=rad_sph,
                      w = wspan/hbar_eVs,
                      n=1.0, 
                      wp=8.959/hbar_eVs, 
                      eps_inf=9.695, 
                      gam_drude=0.073/hbar_eVs
                     )


Cabs = sphere.cross_sections()

plt.plot(1240/wspan, Cabs*10**8)

idx = np.where(Cabs==max(Cabs))
print(1240/wspan[idx])
# T_int = sphere.T_int()    
# print(T_int)
```
